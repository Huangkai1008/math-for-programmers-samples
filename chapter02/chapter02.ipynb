# Chapter 2: Drawing with 2D Vectors

## 2.1 Picturing 2D Vectors

```python
from vectors import *
from vector_drawing import *

dino_vectors = [
    (6, 4),
    (3, 1),
    (1, 2),
    (-1, 5),
    (-2, 5),
    (-3, 4),
    (-4, 4),
    (-5, 3),
    (-5, 2),
    (-2, 2),
    (-5, 1),
    (-4, 0),
    (-2, 1),
    (-1, 0),
    (0, -3),
    (-1, -4),
    (1, -4),
    (2, -3),
    (1, -2),
    (3, -1),
    (5, 1),
]

draw(Points(*dino_vectors))
```

```python
draw(Points(*dino_vectors), Segment((6, 4), (3, 1)))
```

**EXERCISE 2.2**


```python
draw(Arrow((2, -2)))
```

**EXERCISE 2.4**


```python
draw(Points(*dino_vectors), Polygon(*dino_vectors))
```

**EXERCISE 2.5**


```python
draw(
    Points(*[(x, x**2) for x in range(-10, 11)]),
    grid=(1, 10),
    nice_aspect_ratio=False,
)
```

## 2.2 Adding Vectors


```python
dino_vectors2 = [add((2, 1), v) for v in dino_vectors]

draw(
    Points(*dino_vectors, color=blue),
    Points(*dino_vectors2, color=red),
    Polygon(*dino_vectors, color=blue),
    Polygon(*dino_vectors2, color=red),
)
```

```python
p1 = (2, 1)
p2 = (-3, 2)
draw(
    Arrow(p1),
    Arrow(p2),
)
```

```python
draw(
    Arrow(add(p1, p2)),
)
```
